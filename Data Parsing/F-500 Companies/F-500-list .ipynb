```python
from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd
import urllib.request
import re
```

```python
omegalink = 'https://www.fortuneindia.com/fortune-500/company-list/reliance-industries?year=2019'
omegasource = requests.get(omegalink).text
omegasoup = BeautifulSoup(omegasource,'lxml')
```

```python
india = omegasoup.find(class_='f500-list-filter')
finaltable=""
for kilo in india.find_all('li'):
    mike = kilo.find('a',href=True)
    november = mike['href']
    
    masterlink = november
    mastersource = requests.get(masterlink).text
    mastersoup = BeautifulSoup(mastersource,'lxml')
    
    foxtrot = mastersoup.find(class_='company-info-container')
    
    rank = foxtrot.find(class_='sl-num')
    rank = ((rank.text)[0:1])

    cname = foxtrot.find(class_='company-header')
    cname = cname['data-company-name']

    industry = foxtrot.find(class_='company-industry')
    industryA = industry.find(class_='industry')
    industryA = (((industryA.text).replace('\n','')).replace('Industry : ',''))
    incoryear = industry.find(class_='inc-year')
    incoryearA = (((incoryear.text).replace('\n','')).replace('Incorporation Year : ',''))
    industry_fin = industryA +","+ incoryearA

    delta = foxtrot.find(class_= 'company-parameters')
    string=""
    for bravo in delta.find_all(class_='table-body'):
        for charlie in bravo.find_all('td'):
            string = string +"-+-"+ charlie.text
    strings=(((((((((((string.replace(',','').replace('-+-',',')).replace('Net Operating Income','')).replace('Profit','')
         ).replace('Assets','')).replace('Net Worth','')).replace('Net Worth','')).replace('Equity Dividend','')).replace('Employee Cost',''))
          .replace('Revenue','')).replace(',,',','))) 

    echo = foxtrot.find(class_= 'profit-ratios')
    strung = ""
    for golf in echo.find_all(class_='table-body'):
        for hotel in golf.find_all('td'):
            strung = strung +"-+-"+ (hotel).text
    strungs= (((((strung.replace('Assets','')).replace('Net Worth','')).replace('Revenue','')).replace('-+-',','))).replace(',,,','')


    row = cname +","+ industry_fin + strings + strungs +"\n"
    finaltable = finaltable + row


print(finaltable)
    
    
    
    
```

```python
csvname = "f500.csv"
j=[0,1]
for i in j:
    file=open(csvname,"wb")
    file.write(bytes(finaltable,encoding="ascii",errors="ignore"))
```
